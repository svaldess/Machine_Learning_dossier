```
import pandas as pd

df = pd.read_csv("data.csv")
df.head()
```

```
df.columns
```

```
df['Offense'] = df['Offense Type'] + '-' + df['Offense Subtype']
df['Offense'].value_counts()
```

```
df['New Offense'] = df['New Offense Type'] + '-' + df['New Offense Sub Type']
df['New Offense'].value_counts()
```

```
drop_columns = ['Recidivism Reporting Year','Offense Classification','Offense Type','Offense Subtype',
                'New Offense Classification','New Offense Type', 'New Offense Sub Type', 'Target Population']
df = df.drop(drop_columns, axis=1)
df.columns
```

```
df['New Offense'].isnull().sum()
df['New Offense'] = df['New Offense'].fillna('No recidivism')
df['New Offense'].value_counts()
```

```
labels_to_drop = ['Other-Animals','Property-Stolen Property','Property-Other Property']
df = df[df['New Offense'].isin(labels_to_drop)  == False ]
```

```
target = 'New Offense'
```

```
from sklearn.model_selection import train_test_split
df_train, df_test = train_test_split(df, train_size=0.8, random_state=42, shuffle=True, stratify=df[target])
```

```
df_test = df_test.drop(target, axis=1)
```

```
print(df_train.shape)
print(df_test.shape)
```

```
df_train.to_csv("./train.csv", index=False)
df_test.to_csv("./test.csv", index=False)
```
