```
import pandas as pd
data = pd.read_csv('./weather.csv', header=0)
```

```
# split the dataset 
from sklearn.model_selection import train_test_split
train_set, test_set = train_test_split(data, test_size=0.2, random_state=42)
```

```
# Define target
#target = 'Temperature (C)'
target = 'Summary'
```

```
# drop target
test_set = test_set.drop(target, axis=1)
```

```
# check uncommon columns
import numpy as np
np.setdiff1d(train_set.columns, test_set.columns)
```

```
# write csv
train_set.to_csv('./weather_train.csv', index=False)
test_set.to_csv('./weather_test.csv', index=False)
```
